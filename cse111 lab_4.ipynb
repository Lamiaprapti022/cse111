```python
class Student:
    def __init__(self, name, cgp, credit=0, department="CSE"):
        self.name = name
        self.cgpa = cgp
        self.credits = credit
        self.department = department
        self.scholarship = "No scholarship"

    def checkScholarshipEligibility(self):
        if self.cgpa >= 3.7:
            self.scholarship = "Merit-based scholarship"
        elif self.cgpa >= 3.5 and self.credits > 10:
            self.scholarship = "Need-based scholarship"

    def showDetails(self):
        print(f"Name: {self.name}")
        print(f"Department: {self.department}")
        print(f"CGPA: {self.cgpa}")
        print(f"Number of Credits: {self.credits}")
        print(f"Scholarship Status: {self.scholarship}")
print('--------------------------')
std1 = Student("Alif", 3.99, 12)
print('--------------------------')
std1.checkScholarshipEligibility()
print('--------------------------')
std1.showDetails()
print('--------------------------')
std2 = Student("Mim", 3.4)
std3 = Student("Henry", 3.5, 15,"BBA")
print('--------------------------')
std2.checkScholarshipEligibility()
print('--------------------------')
std3.checkScholarshipEligibility()
print('--------------------------')
std2.showDetails()
print('--------------------------')
std3.showDetails()
print('--------------------------')
std4 = Student("Bob", 4.0, 6, "CSE")
print('--------------------------')
std4.checkScholarshipEligibility()
print('--------------------------')
std4.showDetails()
```

```python
menu = {
    'Chicken Lollipop': 15,
    'Beef Nugget': 20,
    'Americano': 180,
    'Red Velvet': 150,
    'Prawn Tempura': 80,
    'Saute Veg': 200
}

class Foodie:
    def __init__(self, name):
        self.name = name
        self.lists= []
        self.t_spent = 0
        self.waiters_tips = 0

    def show_orders(self):
        if len(self.lists) == 0:
            return f"{self.name} has 0 item(s) in the cart.\nItems: []\nTotal spent: {self.t_spent}."
        else:
            return f"{self.name} has {len(self.lists)} item(s) in the cart.\nItems: {self.lists}\nTotal spent: {self.t_spent}."

    def order(self, *items):
        for i in items:
            foods, quantity = i.split('-')
            quantity = int(quantity)
            if foods in menu:
                self.lists.append(foods)
                total_price = menu[foods] * quantity
                self.t_spent += total_price
                print(f"Ordered - {foods}, quantity - {quantity}, price (per Unit) - {menu[foods]}. Total price - {total_price}")

    def pay_tips(self, tips=0):
        if tips > 0:
            self.waiters_tips += tips
            print(f"Gives {tips}/- tips to the waiter.")

    def show_orders(self):
        if len(self.lists) == 0:
            return f"{self.name} has 0 item(s) in the cart.\nItems: []\nTotal spent: {self.t_spent}."
        else:
            return f"{self.name} has {len(self.lists)} item(s) in the cart.\nItems: {self.lists}\nTotal spent: {self.t_spent + self.waiters_tips}"


f1 = Foodie('Frodo')
print(f1.show_orders())
print('1----------------------')
f1.order('Chicken Lollipop-3', 'Beef Nugget-6', 'Americano-1')
print('2----------------------')
print(f1.show_orders())
print('3----------------------')
f1.order('Red Velvet-1')
print('4----------------------')
f1.pay_tips(20)
print('5----------------------')
print(f1.show_orders())

f2 = Foodie('Bilbo')
print('6----------------------')
f2.order('Prawn Tempura-6', 'Saute Veg-1')
print('7----------------------')
f2.pay_tips()
print('8----------------------')
print(f2.show_orders())
```

```python
class Shopidify:
    def __init__(self, name="Guest"):
        self.name = name
        self.cart = {}
        self.purchase_history = []
        print(f"Welcome to Shopidify {name}")

    def add_to_cart(self, *items):
        if self.name != "Guest":
            for item in items:
                if item in self.cart:
                    self.cart[item] += 1
                else:
                    self.cart[item] = 1
        else:
            print(f"Guest cannot add items to the cart.")

    def display_cart(self):
        if self.cart:
            print(f"Items in the cart for {self.name}:")
            for item, quantity in self.cart.items():
                print(f"- {item}: {quantity}x")
        else:
            print(f"{self.name} has no items in the cart.")

    def checkout(self):
        if self.name != "Guest":
            if self.cart:
                self.purchase_history.append(dict(self.cart))
                self.cart.clear()
                print(f"Checkout completed for {self.name}")
            else:
                print(f"{self.name} has no items in the cart to checkout.")
        else:
            print("Guest cannot checkout.")

    def display_history(self):
        if self.name != "Guest":
            if self.purchase_history:
                print(f"Purchase history for {self.name}:")
                for i, transaction in enumerate(self.purchase_history, 1):
                    print(f"Transaction {i}:")
                    for item, quantity in transaction.items():
                        print(f"- {item}: {quantity}x")
            else:
                print(f"{self.name} has no purchase history.")
        else:
            print("Guest has no purchase history.")

# Test the Shopidify class using the provided driver code
guest_account = Shopidify()
print("1xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account = Shopidify("John")
print("2xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan", 2)
guest_account.add_to_cart("Luffy Action Figure")
guest_account.display_cart()
print("3xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.add_to_cart("Chocolate Chip Cookies", "Goku Action Figure", "Dumbbells-5kg")
john_account.add_to_cart("Chocolate Chip Cookies")
john_account.display_cart()
print("4xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.add_to_cart("Air Jordan")
guest_account.display_cart()
print("5xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.checkout()
print("6xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
guest_account.display_history()
print("7xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.checkout()
print("8xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")
john_account.display_history()
print("9xxxxxxxxxxxxxxxxxxxxxxxxxxxxxxxx")

```
